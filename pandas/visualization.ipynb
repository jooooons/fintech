## Line Chart
* plot()
* 데이터의 추세

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]
plt.plot(x,y)
y1 = [6,7,8,9,10]
plt.plot(x,y1)
plt.savefig('./pandas_data/test_vis.jpg')
```

```python
x1 = range(50)
y1 = [np.cos(value) for value in x1]

x2 = range(50)
y2 = [np.sin(value) for value in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, color='red', linestyle='dotted')

plt.subplot(1, 2, 2)
plt.plot(x2, y2, color = 'blue', linestyle='dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1,y1, c='#a52a2a', ls='-.', label='Line1')
plt.plot(x2,y2, c='#0000ff', ls=':', label='Line2')

plt.text(50,60, 'Line 1')
plt.annotate('Line 2', xytext = (30, 110), xy = (50,100), arrowprops=dict(facecolor='yellow', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc='upper right')

plt.xlabel('X_axis')
plt.ylabel('Y_axis')
```

```python
df = pd.read_csv('./pandas_data/seoul_weather.csv', encoding='cp949', skiprows=11)
df
```

```python
df.columns
```

```python
df = df.drop(['\t\t지점번호', '지점명', '\t최고기온시각', '최저기온시각'], axis=1)
df.head()
```

```python
df.columns=['일시', '평온', '고온', '저온']
df.head()
```

```python
df.info()
```

```python
df['일시'] = pd.to_datetime(df['일시'])
df.info()
```

```python
# df['일시'].dt.year/month/day/dayofweek/...
```

```python
# 8월의 최고기온을 추출해서 히스토그램을 그리시오
df[df['일시'].dt.month == 8]['고온'].plot(kind='hist', bins=20)
```

```python
# 매년 8월 중의 최고기온 하나를 line chart로 그리시오
mon8 = df.loc[df['일시'].dt.month == 8]
mon8['year'] = mon8['일시'].dt.year
mon8.groupby('year')['고온'].max().plot()
```

```python
# 지구온난화 여부 확인을 위해 매년 연평균기온의 line chart를 그리시오
df['year'] = df['일시'].dt.year
df.groupby('year')['평온'].mean().plot()
```

```python
# 1 ~ 12 각월의 일평균기온의 box chart를 그리시오
weather_list = []

for i in range(1,13):
    weather_list.append([k for j,k in zip(df['일시'],df['평온']) if j.month == i])
weather_list
```

```python
len(weather_list)
```

```python
pd.DataFrame(weather_list).T.plot(kind='box')
```
