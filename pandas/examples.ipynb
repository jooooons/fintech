> 서울시 구별 CCTV 현황 분석
* 구별 인구당 cctv 댓수 확인
* 인구당 cctv 수와 상관관계가 높은 요인

```python
cctv_seoul = pd.read_csv('./pandas_data/01. CCTV_in_Seoul.csv')
cctv_seoul.head()
```

```python
cctv_seoul.rename(columns={'기관명':'구별'},inplace=True)
cctv_seoul.head()
```

```python
pop_seoul = pd.read_excel('./pandas_data/01. population_in_Seoul.xls', header=2, usecols=[1,3,6,9,13])
pop_seoul.head()
```

```python
pop_seoul.columns = ['구별', '인구수', '한국인', '외국인', '고령자']
pop_seoul.head()
```

```python
cctv_seoul.head(3)
```

```python
pop_seoul.head(3)
```

```python
# CCTV 수를 기준으로 구를 정렬하시오..
cctv_seoul.sort_values('소계')
```

```python
# 2016년도에 cctv 수 증가율이 높은 상위 5개 구..
cctv_seoul['16증가율'] = (cctv_seoul['2016년'] - cctv_seoul['2015년']) / cctv_seoul['2015년']
cctv_seoul.sort_values('16증가율', ascending=False).iloc[:5]
```

```python
# 인구당 cctv 수가 가장 높은 구 상위 5개..
merged_df = pd.merge(cctv_seoul, pop_seoul, on='구별')
merged_df['인구당cctv수'] = merged_df['소계']/merged_df['인구수']
merged_df.sort_values('인구당cctv수', ascending=False).iloc[:5]
```

```python
# # 인구수, 외국인비율, 고령자 비율 중에서 인구당 cctv 수에 가장 큰 영향을 미치는 요소는 무엇인가..?
# merged_df['외국인비율'] = merged_df['외국인']
# merged_df[['인구당cctv수', '인구수', '외국인비율', '고령자비율']].corr()
```

```python
# 출근시간에 승차인원이 가장 많은 역과 하차 인원이 가장 많은 역은? (7시 ~ 9시)
df['morning_in'] = df.iloc[1:,10].apply(lambda x : int(x)) + df.iloc[1:,12].apply(lambda x : int(x))
in_max_people = df['morning_in'].sort_values(ascending=False).max()
df[df['morning_in'] == in_max_people]['지하철역']
```

```python
df['morning_out'] = df.iloc[1:,11].apply(lambda x : int(x)) + df.iloc[1:,13].apply(lambda x : int(x))
out_max_people = df['morning_out'].sort_values(ascending=False).max()
df[df['morning_out'] == out_max_people]['지하철역']
```
